```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(1)

df0 = pd.DataFrame(np.random.normal(5, 1, size=(250, 150)))
df1 = pd.DataFrame(np.random.normal(15, 1, size=(250, 150)))


dataset = pd.concat([df0, df1])
df = dataset.sample(frac=1)
print(df)


plt.scatter(df[0], df[1])
plt.show()

np.savetxt('Gaussian Distribution Data Set.txt', df)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(1)

df0 = pd.DataFrame(np.random.normal(5, 1, size=(250, 150)))
df1 = pd.DataFrame(np.random.normal(15, 1, size=(250, 150)))



dataset = pd.concat([df0, df1])
mu, std = 0, 0.05 * np.std(dataset)
print(std)
noise = np.random.normal(mu, std, size = dataset.shape)
df = dataset.sample(frac=1) + noise
print(df)

plt.scatter(df[0], df[1])
plt.show()

np.savetxt('Gaussian Distribution Data Set.txt', df)
```
